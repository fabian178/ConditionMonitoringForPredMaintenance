```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import grid
import pandas as pd
import pickle
import scipy
from scipy import signal
from scipy.signal import lfilter, lfilter_zi, filtfilt, butter
import pylab
from sklearn.cluster import KMeans
import scipy.io
from scipy.io import *
import os, sys
import warnings
import random
from tkinter import *
import glob
import soundfile as sf #to install use cmd -> "conda install cffi numpy" -> "pip install conda; conda init" -> "pip install pysoundfile"

from tkinter import filedialog
from tkinter import *
from sklearn.externals import joblib
#-------------------------------------------------------------------------------------------------------------------------
def run_once(f):
    def wrapper(*args, **kwargs):
        if not wrapper.has_run:
            wrapper.has_run = True
            return f(*args, **kwargs)
    wrapper.has_run = False
    return wrapper

#run once für einmalige initialisierung der Variablen

@run_once
def initialisieren():
    global var21, var22
    var21 = IntVar()
    var22 = IntVar()
    global var31,var32,var301,var302,var3001,var3002
    var31 = IntVar()
    var32 = IntVar()
    var301 = IntVar()
    var302 = IntVar()
    var3001 = IntVar()
    var3002 = IntVar()
    global var41,var42,var401,var402
    var41 = IntVar()
    var42 = IntVar()
    var401 = IntVar()
    var402 = IntVar()
    global var51,var52,var501,var502,var503,var5001,var5002
    var51 = IntVar()
    var52 = IntVar()
    var501 = IntVar()
    var502 = IntVar()
    var503 = IntVar()
    var5001 = IntVar()
    var5002 = IntVar()
    
master = Tk()
master.title("Select Option")

def weiter():
    run_once(initialisieren())
    global var21, var22
    global var31,var32,var301,var302,var3001,var3002
    global var41,var42,var401,var402
    global var51,var52,var501,var502,var503,var5001,var5002
    
    if var1.get() == 1 and var2.get() == 0 and var3.get() == 0 and var4.get() == 0 and var5.get() == 0:
        #var1 = IntVar()
        #Checkbutton(master, text="Backround", variable=var1).grid(row=1,column=0, sticky=W
        print("Läuft noch nicht!")
    
    if var2.get() == 1 and var3.get() == 0 and var4.get() == 0 and var1.get() == 0 and var5.get() == 0:
        var21 = IntVar()
        Checkbutton(master, text="gut_gespannt", variable=var21).grid(row=2,column=1, sticky=W)
        var22 = IntVar()
        Checkbutton(master, text="schlecht_gespannt", variable=var22).grid(row=3,column=1, sticky=W)
  
    if var3.get() == 1 and var2.get() == 0 and var4.get() == 0 and var1.get() == 0 and var5.get() == 0:
        var31 = IntVar()
        Checkbutton(master, text="Werkstück_1", variable=var31).grid(row=4,column=1, sticky=W)
        var32 = IntVar()
        Checkbutton(master, text="Werkstück_2", variable=var32).grid(row=5,column=1, sticky=W)
        var301 = IntVar()
        Checkbutton(master, text="Scheibe_scharf", variable=var301).grid(row=4,column=2, sticky=W)
        var302 = IntVar()
        Checkbutton(master, text="Scheibe_stumpf", variable=var302).grid(row=5,column=2, sticky=W)
        var3001 = IntVar()
        Checkbutton(master, text="Vorschubverringerung", variable=var3001).grid(row=4,column=3, sticky=W)
        var3002 = IntVar()
        Checkbutton(master, text="keine_Vorschubverringerung", variable=var3002).grid(row=5,column=3, sticky=W)
    
    if var4.get() == 1 and var2.get() == 0 and var3.get() == 0 and var1.get() == 0 and var5.get() == 0:
        var41 = IntVar()
        Checkbutton(master, text="Werkstück 1", variable=var41).grid(row=6,column=1, sticky=W)
        var42 = IntVar()
        Checkbutton(master, text="Werkstück 2", variable=var42).grid(row=7,column=1, sticky=W)
        var401 = IntVar()
        Checkbutton(master, text="Scheibe Scharf", variable=var401).grid(row=6,column=2, sticky=W)
        var402 = IntVar()
        Checkbutton(master, text="Scheibe Stumpf", variable=var402).grid(row=7,column=2, sticky=W)

    if var5.get() == 1 and var2.get() == 0 and var3.get() == 0 and var1.get() == 0 and var4.get() == 0: 
        var51 = IntVar()
        Checkbutton(master, text="Scheibe scharf", variable=var51).grid(row=8,column=1, sticky=W)
        var52 = IntVar()
        Checkbutton(master, text="Scheibe stumpf", variable=var52).grid(row=9,column=1, sticky=W)
        var501 = IntVar()
        Checkbutton(master, text="Nominaldrehzahl", variable=var501).grid(row=8,column=2, sticky=W)
        var502 = IntVar()
        Checkbutton(master, text="Drehzahl -30%", variable=var502).grid(row=9,column=2, sticky=W)
        var503 = IntVar()
        Checkbutton(master, text="Drehzahl +30%", variable=var503).grid(row=10,column=2, sticky=W)
        var5001 = IntVar()
        Checkbutton(master, text="Nominalvorschub", variable=var5001).grid(row=8,column=3, sticky=W)
        var5002 = IntVar()
        Checkbutton(master, text="Vorschub -50%", variable=var5002).grid(row=9,column=3, sticky=W)    
        
    Button(master, text='Close', command=master.destroy).grid(row=10, sticky=W, pady=4)

        
#Sensor Auswahl
var01 = IntVar()
Checkbutton(master, text="SBSS", variable=var01).grid(row=0,column=0, sticky=W)
var02 = IntVar()
Checkbutton(master, text="MARS", variable=var02).grid(row=0,column=1, sticky=W)   

#Backround
var1 = IntVar()
Checkbutton(master, text="Backround", variable=var1).grid(row=1,column=0, sticky=W)    
#Riemenspannung
var2 = IntVar()
Checkbutton(master, text="Riemenspannung", variable=var2).grid(row=2,column=0, sticky=W)
#Spektrum
var3 = IntVar()
Checkbutton(master, text="Spektrum", variable=var3).grid(row=4,column=0, sticky=W)
#Luftschleifen
var4 = IntVar()
Checkbutton(master, text="Luftschleifen", variable=var4).grid(row=6,column=0, sticky=W)
#Abrichten
var5 = IntVar()
Checkbutton(master, text="Abrichten", variable=var5).grid(row=8,column=0, sticky=W)

Button(master, text='Next', command=weiter).grid(row=10, sticky=W, pady=4)
    
mainloop()

#-------------------------------------------------------------------------------------------------------------------------        


#Sensor Auswahl
status_Sensortyp = [["SBSS",var01.get()],["MARS",var02.get()]]
#print(status_Sensortyp)
#print("\n")

#Riemenspannung
status_Riemenspannung=[["Riemenspannung",var2.get()],["gut_gespannt",var21.get()],["schlecht_gespannt",var22.get()]]
#print(status_Riemenspannung)
#print("\n")

#Spektrum
status_Spektrum=[["Spektrum",var3.get()],["Werkstueck_1",var31.get()],["Werkstueck_2",var32.get()],
                ["Scheibe_scharf",var301.get()],["Scheibe_stumpf",var302.get()],
                ["Vorschubverringerung",var3001.get()],["Keine_Vorschubverringerung",var3002.get()]]
#print(status_Spektrum)
#print("\n")

#Luftschleifen
status_Luftschleifen=[["Luftschleifen",var4.get()],["Werkstueck_1",var41.get()],["Werkstueck_2",var42.get()],
                ["Scheibe_scharf",var401.get()],["Scheibe_stumpf",var402.get()]]
#print(status_Luftschleifen)
#print("\n")

#Abrichten
status_Abrichten=[["Abrichten",var5.get()],["Scheibe_scharf",var51.get()],["Scheibe_stumpf",var52.get()],
                 ["Nominaldrehzahl",var501.get()],["Drehzahl_minus_30_Prozent",var502.get()],["Drehzahl_plus_30_Prozent",var503.get()],
                 ["Nominalvorschub",var5001.get()],["Vorschub_minus_50",var5002.get()]]
#print(status_Abrichten)

#---------------------------------------------------------------------------------------------------------------------

status_list_temp = [status_Riemenspannung,status_Spektrum,status_Luftschleifen,status_Abrichten]

for i in range(len(status_list_temp)):
    if status_list_temp[i][0][1] == 1:
        #print(status_list_temp[i])
        
        if status_Sensortyp[0][1] == 1:
            if status_Sensortyp[1][1] == 1:
                status_list=status_list_temp[i]+status_Sensortyp
            else:
                #print("SBSS")
                status_list=status_list_temp[i]+[status_Sensortyp[0]]
        elif status_Sensortyp[1][1] == 1:
            #print("MARS")
            status_list=status_list_temp[i]+[status_Sensortyp[1]]
        else:
            print("ERROR - Bitte wählen Sie min. einen Sensortypen aus")
            break
#---------------------------------------------------------------------------------------------------------------------            
global status_df
status_df = pd.DataFrame(status_list,columns=["lable","status"])
status_df

def dataframe(AnzPfad,dtype):
    
    temp_df = [0,0,0,0,0,0]
    if dtype == "mat":
            df = pd.DataFrame(columns = ["SBSS","Position","Trigger","BMA"])
    elif dtype == "wav":
            df = pd.DataFrame(columns = ["MARS Mic 1","MARS Mic 2","Position","Trigger","Samplerate"])
    else:
            return("*** read dtype ERROR in function dataframe() !***")
#---------------------------------------------------------------------------------------------------------------------    
    for i in range(len(AnzPfad)):
        path = path_folder + "\\" + AnzPfad[i]
        
        if dtype == "mat":
            data_temp = scipy.io.loadmat(path)
            data_temp = np.array(data_temp['raw'])
            data_temp = data_temp.flatten()
        elif dtype == "wav":
            data, samplerate = sf.read(path)
            data_temp = data
        else:
            return("*** read dtype ERROR in function dataframe() !***")
#---------------------------------------------------------------------------------------------------------------------
        if(path.find('J1') != -1 or path.find('J2') != -1 or path.find('J3') != -1 or path.find('J4') != -1):
            if(path.find('J1') != -1):
                index_pos = path.index('J1')
                temp_df[1]=(path[index_pos:index_pos+2])
            elif(path.find('J2') != -1):
                index_pos = path.index('J2')
                temp_df[1]=(path[index_pos:index_pos+2])
            elif(path.find('J3') != -1):
                index_pos = path.index('J3')
                temp_df[1]=(path[index_pos:index_pos+2])
            elif(path.find('J4') != -1):
                index_pos = path.index('J4')
                temp_df[1]=(path[index_pos:index_pos+2])
#---------------------------------------------------------------------------------------------------------------------
            if(path.find('bma') != -1):
               temp_df[3]= 1
            else:
               temp_df[3] = 0

            if(path.find('Trigger1') != -1):
               temp_df[2]= 1
            elif(path.find('Trigger2') != -1):
               temp_df[2] = 2
            elif(path.find('Trigger3') != -1):
               temp_df[2] = 3
#---------------------------------------------------------------------------------------------------------------------
        if (dtype == "mat"):
            df.set_value(i+1, ('SBSS'), data_temp)
            df.set_value(i+1, 'Position', temp_df[1])
            df.set_value(i+1, 'Trigger', temp_df[2])
            df.set_value(i+1, 'BMA', temp_df[3])
            pd.concat([df], axis=1)
        elif (dtype == "wav"):
            df.set_value(i+1, ('MARS Mic 1'), data_temp[0])
            df.set_value(i+1, ('MARS Mic 2'), data_temp[1])
            df.set_value(i+1, 'Position', temp_df[1])
            df.set_value(i+1, 'Trigger', temp_df[2])
            df.set_value(i+1, 'Samplerate', samplerate)  
            pd.concat([df], axis=1)
        else:
            return("*** crating df ERROR in function dataframe() !***")
        
    return df
#---------------------------------------------------------------------------------------------------------------------

# Diese Funktion liest die Daten ein und gibt ein pd.DataFrame zurück den jeweiligen Dateipfaden

path_folder = "F:\\AMELI_KW9\\Messdaten_SM_KW9"
dir_folder = os.listdir(path_folder) 
#print (dir_folder)


status_df = status_df[status_df.status != 0]     #löscht alles Zeilen aus Dataframe, die in der Spalte status keine 1 haben!
#print(status_df)

if(status_df.iloc[-1]["lable"] == "MARS" and status_df.iloc[-2]["lable"] == "SBSS") or (status_df.iloc[-2]["lable"] == "MARS" and status_df.iloc[-1]["lable"] == "SBSS"):
    minus = 2
elif (status_df.iloc[-1]["lable"] == "MARS" or "SBSS"):
    minus = 1
else: minus = 0

for i in range(status_df.shape[0]-minus):
    for index in range(len(dir_folder)):
        if status_df.iloc[i][0] in dir_folder[index]:
            #print("Der index ist: %d" % index)
            nr = index
            break
         
    path_folder = path_folder +"\\"+ dir_folder[nr]
    dir_folder = os.listdir(path_folder)   
    #print(dir_folder)

#print(len(dir_folder))


if(status_df.iloc[-1]["lable"] == "MARS" and minus == 1):
    #print("Mars")
    wavCounter = glob.glob1(path_folder,"*.wav")
    df = dataframe(wavCounter,"wav")
elif(status_df.iloc[-1]["lable"] == "SBSS" and minus == 1):
    #Zählt die Anzahl der MAT-Dateien
    matCounter = glob.glob1(path_folder,"*.mat")
    df = dataframe(matCounter,"mat")
else:
    matCounter = glob.glob1(path_folder,"*.mat")
    df = dataframe(matCounter,"mat")
    wavCounter = glob.glob1(path_folder,"*.wav")
    df = pd.concat([df,dataframe(wavCounter,"wav")], axis=1)
    
global df

#---------------------------------------------------------------------------------------------------------------------

#Zum Speichern des DataFrames

master = Tk()
master.title("Speichern unter...")

def save():
    master.destroy()
    
    filename = ""
    for i in range(len(status_df)):
        filename += status_df.iloc[i]["lable"]
        if (i < len(status_df)-1):
            filename += "-"
    
    root = Tk()
    root.filename =  filedialog.asksaveasfilename(initialfile=filename,initialdir = "/",title = "Select file",filetypes = (("joblib","*."),("all files","*.*")))
    global save_as_path
    save_as_path = root.filename
    root.destroy()
    
    joblib.dump(df,save_as_path)
    print("- saving succesfull -")

Label(master, text="Wollen Sie den DataFrame speichern ?",fg = "black",font = "Verdana 14 bold").pack()
Button(master, text='YES',font = "Verdana 14 bold",width=25, command=save).pack()
Button(master, text='Cancel',font = "Verdana 14 bold",width=25, command=master.destroy).pack()
    
mainloop()

print(df)

#----------------------------------------------------------------------------------------
"""
#zum öffnen der joblib datei...
print(save_as_path)
#save_as_path = "C://Users//wehmbenj//Desktop//Riemenspannung-gut_gespannt-SBSS2"
input_df = joblib.load(save_as_path)
print(input_df)
"""
```
